```python
import numpy as np
import cv2
import PIL.Image as Image
import os
import matplotlib.pylab as plt
import tensorflow as tf
import tensorflow_hub as hub
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
img_shape=(224,224)
classifier = tf.keras.Sequential([
    hub.KerasLayer("https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/4", input_shape=img_shape+(3,))
])
classifier.trainable=False
```

```python
image=Image.open("/content/drive/MyDrive/Colab Notebooks/ppp/flower_photos/tulips/113902743_8f537f769b_n.jpg").resize(img_shape)
plt.imshow(image)
```

```python
flower=np.array(image)/255
flower=flower[np.newaxis, ...]

result=classifier.predict(flower)
Max_result=np.argmax(result)

image_lables=[]
with open("/content/drive/MyDrive/Colab Notebooks/ppp/flower_photos/LICENSE.txt") as f:
  image_labels=f.read().splitlines()
image_labels[Max_result]
```

```python
flower_folder=r"/content/drive/MyDrive/Colab Notebooks/ppp/flower_photos"
catagories=['daisy','dandelion','roses','sunflowers','tulips',]
```

```python
data=[]
for Catagories in catagories:
  Folder=os.path.join(flower_folder,Catagories)
  label=catagories.index(Catagories)

  for img in os.listdir(Folder):
    image=os.path.join(Folder,img)
    image_arr=cv2.imread(image)
    image_arr=cv2.resize(image_arr,img_shape)
    data.append([image_arr,label])
```

```python
data
```

```python
import random
random.shuffle(data)
x=[]
y=[]
for feature,label in data :
  x.append(feature)
  y.append(label)

x=np.array(x)
y=np.array(y)

x=x/255
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
previous_model=hub.KerasLayer(
    "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4",
    input_shape=img_shape+(3,),
    trainable=False
)
```

```python
number_of_flower=5
model = tf.keras.Sequential([
                 previous_model,
                 tf.keras.layers.Dense(number_of_flower)          
])
model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc']
)
model.fit(xtrain,ytrain,epochs=5)
```

```python
model.evaluate(xtest,ytest)
```
